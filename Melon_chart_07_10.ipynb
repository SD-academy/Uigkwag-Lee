```python
Okt_temp = []
Kkma_temp = []
for i in range(len(lyrics)):
  if len(lyrics[i]) != 0:
    Okt_temp.append(Okt.pos(lyrics[i]))
    Kkma_temp.append(Kkma.pos(lyrics[i]))
  else:
    Okt_temp.append(' ')
    Kkma_temp.append(' ')
```

```python
Okt_temp[0]
```

```python
def Okt_Kkma_nouns(p_pos,p_nouns):
  p_pos_temp = []
  for i in p_pos:
    temp = []
    for j in i:
      if len(j) != 1 :
        if j[1] == p_nouns:
          if len(j[0]) != 1:
            temp.append(j[0])
      else:
        temp.append(' ')
    p_pos_temp.append(temp)
  
  return p_pos_temp

Okt_temp = Okt_Kkma_nouns(Okt_temp,'Noun')
Kkma_temp = Okt_Kkma_nouns(Kkma_temp,'NNG')
```

```python
# TF-IDF로 단어 가중치 계산
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
```

```python
tdm = tf.fit_transform(temp)
```

```python
word_count = pd.DataFrame({
    'word': tf.get_feature_names(),
    'frequency': tdm.sum(axis=0).flat
})
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tdm,tdm)
```

```python
indices = pd.Series(df_drop.index,index=df_drop['title'])
```

```python
indices.head()
```

```python
idx = indices['잠이 오질 않네요']
```

```python
print(idx)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]

    music_indices = sorted([i[0] for i in sim_scores])

    for i in range(len(music_indices)):
      try:
        if music_indices[i] > len(df_drop):
            del music_indices[i:]
      except:
        break

    return df_drop['title'].iloc[music_indices]
# 협업 필터링
```
