```python
# TF-IDF로 단어 가중치 계산
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
```

```python
tdm = tf.fit_transform(temp)
```

```python
word_count = pd.DataFrame({
    'word': tf.get_feature_names(),
    'frequency': tdm.sum(axis=0).flat
})
```

```python
word_count.sort_values('frequency',ascending=False).head(20)
#전체 달에 관한 단어별 빈도
```

```python
tdm.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tdm,tdm)
```

```python
indices = pd.Series(df_drop.index,index=df_drop['title'])
```

```python
indices.head()
```

```python
idx = indices['Lovesick Girls']
```

```python
print(idx)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]

    music_indices = sorted([i[0] for i in sim_scores])

    for i in range(len(music_indices)):
      try:
        if music_indices[i] > len(df_drop):
            del music_indices[i:]
      except:
        break

    return df_drop['title'].iloc[music_indices]
# 협업 필터링
```
